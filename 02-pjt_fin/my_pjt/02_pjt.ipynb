```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import copy
cf = pd.read_csv('archive/NFLX.csv', encoding='cp949', usecols=['Date', 'Open', 'High', 'Low', 'Close' ])
df = copy.deepcopy(cf)
```

```python
idxs = df[df['Date'] <= "2021-01-01"].index
df.drop(idxs, inplace=True)

# df = df[df['Date'].between("2021-01-01","2024-01-01") ]

# to_datatime()

df["Date"] = pd.to_datetime(df["Date"])

plt.plot(df['Date'], df['Close'])


#   
plt.title('NFLX Close Price')

# x  
plt.xlabel('Date')

# y  
plt.ylabel('Close Price')

# x  ()
plt.xticks(rotation=45)

#  
plt.show()
```

```python
df.describe()
df

max_price = max(df.loc[:, 'Close'])
min_price = min(df.loc[:, 'Close'])


print(" " ,max_price)
print(" " ,min_price)
```

```python
datetimeIndex = pd.DatetimeIndex(df['Date'])
idx = pd.to_datetime(datetimeIndex)
midx = idx.to_period("M")
ef = df.groupby(midx).mean(numeric_only=True).reset_index()

ef['Date'] = ef['Date'].astype(str)

plt.plot(ef['Date'], ef['Close'])


#   
plt.title('Monthly Average Close Price')

# x  
plt.xlabel('Date')

xticks = [date for date in ef['Date'] if int(date.split('-')[1]) % 2 != 0]
plt.xticks(xticks, rotation=45)

# y  
plt.ylabel('Avarage Close Price')
# 
plt.show()
```
